## 【設問1】
### データ分析では、データが欠損していること（一部のデータが欠けていること）があります。欠損したままのデータを処理すると、たとえば、平均値などの結果に正しい値が得られなくなります。そのため、欠損値は除外することが必要になります。
### 欠損値は、numpyモジュールにおけるnumpy.nanという特別な値で表現することができます。
### 課題２で使用したDataFrame型のオブジェクトattr1_dataframe1を使って、データの欠損について学びます。




```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

attr1 = {'ID':['100', '102', '104', '106', '108'],
         'City':['Tokyo', 'Osaka', 'Kyoto', 'Hokkaido', 'Tokyo'],
         'Birth_year':[1990, 1989, 1992, 1997, 1982],
         'Name':['Hiroshi', 'Akiko', 'Yuki', 'Satoru', 'Steeve']}

attr1_dataframe1 = DataFrame(attr1)
attr1_dataframe1         
```

### 【設問１－１】解答
#### データattr1_dataframe1のすべての要素の中に、Tokyoという文字列が存在するかどうかをisinメソッドを使って調べてください。
#### isinを用いると、要素にTokyoが含まれているとき、Trueを表示し、要素にTokyoが含まれていないとき、Falseを表示します。

```python
attr1_dataframe1.isin(['Tokyo'])
```

### 【設問１－２】解答
#### データattr1_dataframe1の列Nameの値をすべてnanに変更した新しいデータattr1_dataframe2を作成し、画面に表示させてください。

```python
attr1_dataframe1.loc[:,['Name']] = np.nan
attr1_dataframe1
```

#### データの要素がnanであるかどうかは、isnull()メソッドによって判定することができます。

### Pandasを使ってcsvファイルを読み込み、統計量を計算しましょう。

### 【設問３－１】解答
#### titanic.csvをDataFrame型データに読み込み、先頭の５行を表示させてください。

```python
import pandas as pd
df_titanic = pd.read_csv('titanic.csv').head(5)
df_titanic
```

### 【設問３－２】解答
#### Titanic号の乗員の年齢構成を調べ、
+ 9歳以下　　 　〇〇人
+ 10歳から19歳　〇〇人
* 20歳から29歳　〇〇人
* 30歳から39歳　〇〇人
+ 40歳から49歳　〇〇人
+ 50歳から59歳　〇〇人
+ 60歳から69歳　〇〇人
+ 70歳以上　　　〇〇人
#### のように表示しなさい。

```python
import pandas as pd
df_titanic1 = pd.read_csv('titanic.csv')


age_9 = 0
age_10_19 = 0
age_20_29 = 0
age_30_39 = 0
age_40_49 = 0
age_50_59 = 0
age_60_69 = 0
age_70 = 0

for a in df_titanic1['Age']:
    if a <= 9 :
        age_9 += 1
    elif a <= 19 :
        age_10_19 += 1
    elif a <= 29 :
        age_20_29 += 1
    elif a <= 39 :
        age_30_39 += 1
    elif a <= 49 :
        age_40_49 += 1
    elif a <= 59 :
        age_50_59 += 1
    elif a <= 69 :
        age_60_69 += 1
    elif a >= 70 :
        age_70 += 1

print('9歳以下:', age_9)
print('10歳から19歳:', age_10_19)
print('20歳から29歳:', age_20_29)
print('30歳から39歳:', age_30_39)
print('40歳から49歳:', age_40_49)
print('50歳から59歳:', age_50_59)
print('60歳から69歳:', age_60_69)
print('70歳以上:', age_70)
```

### 【設問３－３】解答
#### 設問３－２で求めた年齢構成を円グラフで表してください。


```python
series = pd.Series([age_9, age_10_19, age_20_29, age_30_39, age_40_49, age_50_59, age_60_69, age_70], index = ['9<', 
                                                                                                              '10 - 19', 
                                                                                                              '20 - 29', 
                                                                                                              '30 - 39',
                                                                                                              '40 - 49',
                                                                                                              '50 - 59',
                                                                                                              '60 - 69',
                                                                                                              '>70'], name = 'AGE')
series.plot.pie(figsize = (4, 4))
plt.show()
```
